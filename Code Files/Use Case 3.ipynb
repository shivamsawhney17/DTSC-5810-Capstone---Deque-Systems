## Volume Analysis: number of interactions over time to see trends or patterns. For instance, are there more interactions on certain days or times?

```python
interaction_df
```

```python
import matplotlib.pyplot as plt

interaction_df['datetime'] = pd.to_datetime(interaction_df['datetime'])
interaction_df.set_index('datetime', inplace=True)
interaction_volume = interaction_df.resample('D').size()

# Plotting
plt.figure(figsize=(10, 5))
interaction_volume.plot(title='Interaction Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Interactions')
plt.show()
```

## Response Time Analysis: to evaluate the efficiency of response

```python
import re


def extract_seconds(response):
    # Check if the response is a string instance
    if isinstance(response, str):
        match = re.search(r'Time taken for first response: (\d+) seconds', response)
        return int(match.group(1)) if match else None
    return None


# Apply this function to the response column
interaction_df['response_time_seconds'] = interaction_df['response'].apply(extract_seconds)
```

```python
# Plotting the histogram of response times
plt.hist(interaction_df['response_time_seconds'], bins=30, alpha=0.7)
plt.title('Response Time Distribution')
plt.xlabel('Response Time (seconds)')
plt.ylabel('Frequency')
plt.show()
```

## Multilevel Question-Topic analysis -->heirarchy of question_topics

```python

interaction_df[['question_topic', 'question_topic2', 'question_topic_3', 'question_topic4']] = interaction_df[['question_topic', 'question_topic2', 'question_topic_3', 'question_topic4']].applymap(str)

interaction_df['hierarchical_topic'] = interaction_df[['question_topic', 'question_topic2', 'question_topic_3', 'question_topic4']].apply(
    lambda x: ' > '.join([item for item in x if item.lower() not in ['nan', 'false']]), axis=1)

filtered_topics_df = interaction_df[~interaction_df['hierarchical_topic'].isin(['', 'Off topic'])]

# Recalculate the value counts for the filtered DataFrame
filtered_topic_counts = filtered_topics_df['hierarchical_topic'].value_counts().head(10)

# Now plot the bar chart with the filtered data
plt.figure(figsize=(12, 6))
filtered_topic_counts.plot(kind='bar', color='skyblue')
plt.title('Top Hierarchical Question Topics (excluding \'FALSE\' and empty categories)')
plt.xlabel('Hierarchical Topics')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
from textblob import TextBlob

# Assuming 'prompt' and 'response' are the columns with text data
interaction_df['prompt_sentiment'] = interaction_df['prompt'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
interaction_df['response_sentiment'] = interaction_df['response'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

```python
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.hist(interaction_df['response_sentiment'].dropna(), bins=50, color='green', alpha=0.7)
plt.title('Distribution of Response Sentiments')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Keyword Analysis

```python
from sklearn.feature_extraction.text import CountVectorizer

# Ensure all entries in the 'prompt' column are strings
interaction_df['prompt'] = interaction_df['prompt'].astype(str)

# Create a CountVectorizer object to find keyword frequencies, excluding common English stop words
vectorizer = CountVectorizer(stop_words='english')

# Transform the data using CountVectorizer
X = vectorizer.fit_transform(interaction_df['prompt'].dropna())  # Ensure no NaN values

# Convert the matrix to a DataFrame and sum each word's occurrence
keyword_freq = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
keyword_freq = keyword_freq.sum().sort_values(ascending=False).head(20)

# Show the most frequent words
print(keyword_freq)

```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Convert the frequency series into a dictionary for the WordCloud generator
word_freq_dict = keyword_freq.to_dict()

# Create a word cloud object
wordcloud = WordCloud(width = 800, height = 400, background_color ='white')

# Generate a word cloud
wordcloud.generate_from_frequencies(word_freq_dict)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # No axis for neat presentation
plt.show()
```
