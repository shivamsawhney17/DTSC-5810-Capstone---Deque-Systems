# Web Scraping WebAIM

```python
#import the required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Send an HTTP request to the specified URL
url = "https://webaim.org/discussion/mail_archive?sort_by=1"
response = requests.get(url)

# Parse the HTML content of the response using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the discussion topics within the HTML content identified by h3 tags
discussion_topics = soup.find_all('h3')
```

```python
# Iterate over each discussion topic extracted from the HTML content
for topic in discussion_topics:
    # Print the text content of each discussion topic, removing any leading or trailing whitespaces
    print(topic.text.strip())
```

```python
# Function to fetch discussion titles from a given page URL
def fetch_titles_from_page(url):
    """
    Fetches discussion titles from a web page.

    Parameters:
    - url (str): The URL of the web page to fetch titles from.

    Returns:
    - list of str: A list of discussion titles extracted from the web page.
    """
    # Send an HTTP request to the specified URL
    response = requests.get(url)

    # Parse the HTML content of the response
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all discussion titles within h3 tags
    discussion_topics = soup.find_all('h3')

    # Extract the text content of each discussion title and remove leading/trailing whitespaces
    return [topic.text.strip() for topic in discussion_topics]

# Base URL for the discussion archive pages
base_url = "https://webaim.org/discussion/mail_archive?sort_by=1&from={}"

# Initialize a list to store all discussion titles
all_titles = []

# Loop through the pages of the discussion archive
page_size = 50
total_titles = 10708  # Total number of discussion titles
for page_start in range(0, total_titles, page_size):
    # Construct the URL for the current page
    page_url = base_url.format(page_start)

    # Fetch discussion titles from the current page and append them to the list
    titles_on_page = fetch_titles_from_page(page_url)
    all_titles.extend(titles_on_page)

# Print the count of fetched titles
print("Total number of fetched titles:", len(all_titles))

# Print the first few titles to verify
print("First few titles:")
for title in all_titles[:5]:
    print(title)
```

```python
# Create a DataFrame using the fetched discussion titles
df = pd.DataFrame({'title': all_titles})
```
