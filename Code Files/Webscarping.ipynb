```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup

# The URL you want to scrape
url = 'https://webaim.org/discussion/archives'

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the page using Beautiful Soup
soup = BeautifulSoup(response.text, 'html.parser')

# Example: Find and print all 'a' tags (links)
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python

links = [link['href'] for link in soup.find_all('a', href=True) if '/discussion' in link['href']]
# Assuming 'links' is your list
links = links[1:]  
links
```

```python
from bs4 import NavigableString

data = []
base_url = 'https://webaim.org'  # Adjust this to the actual base URL

for link in links:
    # Ensure the link is a complete URL
    if link.startswith('/'):
        link = base_url + link
    sub_response = requests.get(link)
    sub_soup = BeautifulSoup(sub_response.text, 'html.parser')
    for li in sub_soup.find_all('li'):
        for content in li.contents:
            if isinstance(content, NavigableString) and " by " in content:
                email_address = content.split(" by ")[1].strip(' "')
                print(email_address)
                data.append({
                    'URL': url,
                    'Subject': email_address
                })

```

```python
# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df['Email/Name'] = df['Subject'].str.extract(r'(.+?) \(')
df['Date'] = df['Subject'].str.extract(r'\((\w+ \d+, \d{4})')
df['Time'] = df['Subject'].str.extract(r'(\d{1,2}:\d{2}[APM]{2})')
```
