Function which will accept the list of product ids, match them with dqu_products table, create a filtered dataframe which will not consist of any WCAG conformance testing or any any inactive courses - ultimately returning the average duration:

```python
def get_avg_duration(prod, df1):
  wcag_ids= [18, 244, 246, 248] #Creating a list of WCAG ids
  temp_df = df1[df1['id'].isin(prod) & ~df1['id'].isin(wcag_ids)] #Creating a temp_df which will have rows of products that are matched by comparing it to ids from `prod`
  filtered_df = temp_df[temp_df['status'] == "active"] #Filtering the dataframe which will contain only active products
  cnt = 0 #Initializing count and sum variables for further calculations
  sm = 0
  for index, row in filtered_df.iterrows(): #Loop for calculating avg. duration
    if row['minutes'] > 0: #Only calculating non-zero values
      sm=sm+row['minutes']
      cnt+=1
  print(filtered_df)
  if cnt > 0: #Added code to avoid division by zero
    avg = round(sm/cnt, 2)
    print('\nThe Average Duration for this package is: ', avg, "min.")
  else:
    print("\nThe Average Duration for this package is: 0 min.")
```

## Approach 2 - Single function and output in a csv (Easier on your end)

Creating a global dataframe to store the values of package id and the average duration in minutes:

```python
df_avg = pd.DataFrame(columns=['Package id', 'Average Duration (min.)'])
```

A single function which will do all the work - user needs to input package id and will get the resultant average duration in return:

```python
def get_avg_duration_sf(id):
  global df_avg

  temp = []
  prod = []
  wcag_ids= [18, 244, 246, 248]
  cnt = 0
  sm = 0

  for i in range(len(df2)):
    if df2.id[i] == id:
      temp = df2.products[i]
  prod = [int(num) for num in temp.split(',')]

  temp_df = df1[df1['id'].isin(prod) & ~df1['id'].isin(wcag_ids)]
  filtered_df = temp_df[temp_df['status'] == "active"]
  for index, row in filtered_df.iterrows():
    if row['minutes'] > 0:
      sm=sm+row['minutes']
      cnt+=1
  #print(filtered_df)
  if cnt > 0:
    avg = round(sm / cnt, 2)
    if id in df_avg['Package id'].values: #Checking if the id is already present in `df_avg`
      df_avg.loc[df_avg['Package id'] == id, 'Average Duration (min.)'] = avg #If yes, then updating the existing value for duration
      print('\nThe Average Duration for this package is: ', avg, "min.")
    else:
      new_entry = pd.DataFrame({'Package id': [id], 'Average Duration (min.)': [avg]}) #If no, then appending new data entry in the dataframe
      df_avg = pd.concat([df_avg, new_entry], ignore_index=True)
      print('\nThe Average Duration for this package is: ', avg, "min.")
  else: #Case for cnt = 0
    if id not in df_avg['Package id'].values:
      new_entry = pd.DataFrame({'Package id': [id], 'Average Duration (min.)': [0]})
      df_avg = pd.concat([df_avg, new_entry], ignore_index=True)
      print("\nThe Average Duration for this package is: 0 min.")

  print(df_avg)
```
