```python
# list of DataFrames to be concatenated
dfs = [df_18, df_19, df_20, df_21, df_22, df_23, df_24]

# Concatenate the DataFrames in the list into one DataFrame
merged_df = pd.concat(dfs, ignore_index=True)
```

```python
# Column names to be dropped from the merged DataFrame
columns_to_drop = ['TimeZone', 'Status', 'Currency', 'Gross', 'Fee', 'Transaction ID', 'Item ID', 'Sales Tax', 'Option 1 Name', 'Option 1 Value', 'Reference Txn ID', 'Invoice Number', 'Quantity', 'Receipt ID', 'Subject', 'Note', 'Payment Source', 'Custom Number' ]
```

```python
merged_df.drop(columns=columns_to_drop, inplace=True)
```

```python
# Reset the index of the DataFrame, dropping the old index and making the changes in-place
df_cleaned.reset_index(drop=True, inplace=True)
```

```python
df_cleaned.head(10)
```

```python
# Extract the year from the 'Date' column and assign it to a new column 'Year'
df_cleaned['Year'] = df_cleaned['Date'].dt.year

# Extract the month from the 'Date' column and assign it to a new column 'Month'
df_cleaned['Month'] = df_cleaned['Date'].dt.month

# Extract the quarter from the 'Date' column and assign it to a new column 'Quarter'
df_cleaned['Quarter'] = df_cleaned['Date'].dt.quarter
```

## This line of code gets the clean and merged sales dataset.
You can skip all the previous steps and directly work with this clean dataset.

```python
# Save the cleaned DataFrame to a CSV file named 'PayPal_Sales.csv', excluding the index
df_cleaned.to_csv('PayPayl_Sales.csv', index=False)
```

```python
# Calculate yearly revenue by summing the 'Net' column grouped by year
revenue_yearly = df_cleaned.groupby('Year')['Net'].sum()

# Calculate monthly revenue by summing the 'Net' column grouped by year and month
revenue_monthly = df_cleaned.groupby(['Year', 'Month'])['Net'].sum()

# Calculate quarterly revenue by summing the 'Net' column grouped by year and quarter
revenue_quarterly = df_cleaned.groupby(['Year', 'Quarter'])['Net'].sum()
```

```python
print("Yearly online sales revenue:")
print(revenue_yearly)
```

```python
print("\nMonthly online sales revenue:")
print(revenue_monthly.to_string())
```

```python
print("\nQuarterly online sales revenue:")
print(revenue_quarterly.to_string())
```

```python
# Set the style of the seaborn plots to whitegrid
sns.set(style="whitegrid")

# Create a line plot for yearly sales revenue
plt.figure(figsize=(10, 6))
sns.lineplot(data=revenue_yearly, x='Year', y='Net', marker='o', color='blue')
plt.title('Yearly Online Sales Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group the cleaned DataFrame by year and month, sum the 'Net' column, then reset the index
monthly_sales = df_cleaned.groupby(['Year', 'Month'])['Net'].sum().reset_index()

# Set the style of seaborn plots to whitegrid
sns.set(style="whitegrid")

# Create a line plot showing yearly sales by month
plt.figure(figsize=(12, 8))
sns.lineplot(data=monthly_sales, x='Month', y='Net', hue='Year', marker='o', palette='viridis')
plt.title('Yearly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.legend(title='Year')
plt.show()
```

```python
# Group the data by 'Year' and 'Quarter' and calculate total revenue for each group
quarterly_sales = df_cleaned.groupby(['Year', 'Quarter'])['Net'].sum().reset_index()

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a line plot for quarterly sales revenue
plt.figure(figsize=(12, 8))
sns.lineplot(data=quarterly_sales, x='Quarter', y='Net', hue='Year', marker='o', palette='viridis')
plt.title('Quarterly Sales Revenue for Different Years')
plt.xlabel('Quarter')
plt.ylabel('Total Sales Revenue')
plt.legend(title='Year')
plt.show()
```
