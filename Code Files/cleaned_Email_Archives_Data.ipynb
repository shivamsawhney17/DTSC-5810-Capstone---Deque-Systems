```python
for index, row in df.iterrows():
      #selecting the columns which has valid values and are not starting with the term 'Search' in Column_0
    if pd.notna(row['Column_0']) and not row['Column_0'].startswith('Search'):  #notna detects valid values in dataframe
        shifted_row = [None, None]  # Initialize the shifted row with two None values
        shifted_row.extend(row[:-2])  # Add the values from the original row, starting from the third column
        df.iloc[index] = shifted_row  # Assign the shifted row back to the DataFrame
```
